### Q55: Create a heatmap using Year as index and key metrics as columns (normalized values) to visually compare performance across years.

```python
cols = ["market_cap","revenue","earnings","op_margin","eps","total_debt","cash"]
norm = mcd[cols].apply(lambda s: (s - s.min()) / (s.max() - s.min()))
norm["year"] = mcd["year"]
norm = norm.set_index("year")

sns.heatmap(norm, annot=True)
```
