### Q37. Build a performance scorecard table by year with Revenue, Earnings, Operating Margin, EPS, Debt, Cash, Market Cap, and key ratios. 

```python
def label_year(row):
    labels = []
    if row["revenue_g"] > 5:
        labels.append("Growth")
    if row["op_margin"] > mcd["op_margin"].mean():
        labels.append("Margin_Strong")
    if row["net_debt_p"] < mcd["net_debt_p"].shift(1).fillna(row["net_debt_p"]):  
        labels.append("Deleveraging")
    if row["pe"] > mcd["pe"].mean():
        labels.append("Valuation_High")
    return " | ".join(labels) if labels else "Neutral"

mcd["narrative"] = mcd.apply(label_year, axis=1)
mcd[["year","narrative"]]
```

### Q38. Identify the top 3 years to showcase to investors and justify using metrics (growth + profitability + balance sheet).

```python
mcd['investor_score'] = (mcd['revenue_g'].fillna(0).rank(pct=True) * 0.35 +
                         mcd['op_margin'].rank(pct=True) * 0.35 +
                         (-mcd['net_debt_p']).rank(pct=True) * 0.30)
mcd.sort_values('investor_score',ascending=False)[['year','investor_score','revenue_g','op_margin','net_debt_p']].head(3)
```

### Q39. Identify the top 3 risk years and justify using leverage + margin + cash indicators.

```python
mcd['risk_score'] = (mcd['da'].fillna(0).rank(pct=True) * 0.4 +
                         (-mcd['op_margin']).rank(pct=True) * 0.3 +
                         (-mcd['li']).rank(pct=True) * 0.3)
mcd.sort_values('risk_score',ascending=False)[['year','risk_score','da','op_margin','li']].head(3)
```
